```python
max_fuel_cons = data.groupby("aircraft_type")["true_fuel_per_sec"].max().to_dict()
```

```python
train = pd.read_parquet("./data/fuel_train.parquet")
flight_list_train = pd.read_parquet("./data/flightlist_train.parquet")
res = pd.merge(train, flight_list_train,on="flight_id")
```

```python
res["segment_duration"] = (pd.to_datetime(train["end"])-pd.to_datetime(train["start"])).dt.seconds
res["fuel_kg_sec"] = res["fuel_kg"]/res["segment_duration"]
max_fuel_cons = res.groupby("aircraft_type")["fuel_kg_sec"].max().to_dict()
min_fuel_cons = res.groupby("aircraft_type")["fuel_kg_sec"].min().to_dict()
```

```python
data["y_pred"] = data.apply(lambda x: min(x.y_pred, (max_fuel_cons[x.aircraft_type]*x["segment_duration[seconds]"])),axis=1)
```

```python
root_mean_squared_error(y_true=data["y_true"], y_pred=data["y_pred"])
```

```python
data["y_pred"] = data.apply(lambda x: max(x.y_pred, (min_fuel_cons[x.aircraft_type]*x["segment_duration[seconds]"])),axis=1)
```

```python
data_with_vr = data.dropna(subset=["mean_vertical_rate[ft/min]"])
for ac_type in set(data_with_vr["aircraft_type"]):
    print("Aircraft type: "+ac_type)
    temp = data_with_vr[(data_with_vr["aircraft_type"]==ac_type)&(data_with_vr["pred_fuel_per_sec"]>0)&(data_with_vr["pred_fuel_per_sec"]<10)]
    plt.scatter(temp["mean_vertical_rate[ft/min]"],temp["true_fuel_per_sec"], label="true consumption")
    plt.scatter(temp["mean_vertical_rate[ft/min]"],temp["pred_fuel_per_sec"], label="predicted consumption")
    plt.xlabel("mean_vertical_rate")
    plt.ylabel("fuel_per_sec")
    plt.legend()
    plt.show()
```

```python
data_without_traj = data[(data["mean_vertical_rate[ft/min]"].isna())]
for ac_type in set(data_without_traj["aircraft_type"]):
    print("Aircraft type: "+ac_type)
    temp = data_without_traj[(data_without_traj["aircraft_type"]==ac_type)&(data_without_traj["pred_fuel_per_sec"]>0)&(data_without_traj["pred_fuel_per_sec"]<10)]
    plt.scatter(temp["pred_fuel_per_sec"],temp["true_fuel_per_sec"])
    plt.xlabel("pred_fuel_per_sec")
    plt.ylabel("true_fuel_per_sec")
    plt.show()
```
