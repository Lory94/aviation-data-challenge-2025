```python
from prc_challenge.problem import PrcChallenge

problem = PrcChallenge(seed=1)

train_FuelSegment_X = problem.train_FuelSegment_X
train_FuelSegment_Y = problem.train_FuelSegment_Y
train_FlightList = problem.train_FlightList
Flight = problem.Flight
Airport = problem.Airport
column_functions = problem.column_functions
```

```python
from prc_challenge.feature_engineering import AddTimeSinceTakeOff_v0_0_0

step = AddTimeSinceTakeOff_v0_0_0

FuelSegment_X, column_functions = step(
    method="earliest_fuel_segment_start",
)(
    FuelSegment_X=train_FuelSegment_X, 
    FuelSegment_Y=train_FuelSegment_Y, 
    FlightList=train_FlightList, 
    Airport=Airport, 
    Flight=Flight, 
    column_functions=column_functions,
)
```
